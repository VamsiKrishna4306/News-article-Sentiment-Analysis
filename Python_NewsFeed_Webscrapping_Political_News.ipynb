```python
from textblob import TextBlob
import requests
from bs4 import BeautifulSoup
import csv,json
```

```python
class Analysis:
    
    def __init__(self):
        self.subjectivity = 0
        self.sentiment = 0
        
        self.cnn_news_url="http://rss.cnn.com/rss/cnn_allpolitics.rss"
        self.fox_news_url="http://feeds.foxnews.com/foxnews/politics"
    
    
    def scrap_cnn_news (self):
        page = requests.get(self.cnn_news_url)
        soup = BeautifulSoup(page.text,'lxml')
        #print(soup.body.prettify())
        items = soup.body.find_all('item')
        
        csv_file = open("cnn_news.csv",'w+',encoding="utf-8",newline='')
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['published_date','origlink','description','article_text'])
    
        for item in items:
            origlink = item.find('feedburner:origlink').text
            description = item.find('description').text.split('<div')[0]
            published_date = item.find('pubdate').text if not (item.find('pubdate') is None) else 'NO Data'
            print(origlink)
            print(description)
            print(published_date)
            
            article_page = requests.get(origlink)
            article_soup = BeautifulSoup(article_page.text,'lxml')
            article_paragraphs = article_soup.find_all("div", {"class":["zn-body__paragraph", "zn-body__paragraph speakable",
                                                            "el__leafmedia el__leafmedia--sourced-paragraph","article-content rich-text",
                                                                        "article-content--body-text"]})
            article_text = "" 
            for i in article_paragraphs:
                article_text = article_text+i.text+"\n"
            print(article_text)

            csv_writer.writerow([published_date,origlink,description,article_text])
        csv_file.close()
            
    def scrap_fox_news (self):
        page = requests.get(self.fox_news_url)
        soup = BeautifulSoup(page.text,'lxml')
        #print(soup.body.prettify())
        items = soup.body.find_all('item')
        
        csv_file1 = open("fox_news.csv",'w+',encoding="utf-8",newline='')
        csv_writer1 = csv.writer(csv_file1)
        csv_writer1.writerow(['published_date','origlink','description','article_text'])
        
        for item in items:
            origlink = item.find('feedburner:origlink').text
            description = item.find('description').text.split('<img')[0]
            published_date = item.find('pubdate').text if not (item.find('pubdate') is None) else 'NO Data'
            print(origlink)
            print(description)
            print(published_date)
            
            article_page = requests.get(origlink)
            article_soup = BeautifulSoup(article_page.text,'lxml')
            article_body = article_soup.find("div",class_="article-body")
            article_paragraphs = article_body.find_all("p")            
            article_text = "" 
            for i in article_paragraphs:
                article_text = article_text+i.text+"\n"
            print(article_text)
            
            csv_writer1.writerow([published_date,origlink,description,article_text])
        csv_file1.close()
    
a = Analysis()
a.scrap_cnn_news()   
a.scrap_fox_news()  
```

## Sentiment Analysis

```python
from textblob import TextBlob
import pyodbc
import pandas as pd
```

```python
read_connection = pyodbc.connect('Driver={ODBC Driver 13 for SQL Server};Server=tcp:socialmediareader.database.windows.net,\
                     1433;Database=socialmediareader-sql;Uid=vamsi;Pwd=azure@123;Encrypt=yes;\
                     TrustServerCertificate=no;Connection Timeout=30;')
    
cursor_read = read_connection.cursor()
cnn_query = "select * from [dbo].[CNN_NEWS]"
fox_query = "select * from [dbo].[FOX_NEWS]"
CNN_data = pd.read_sql(cnn_query,read_connection) 
FOX_data = pd.read_sql(fox_query,read_connection) 
CNN_data
```

## FOX NEWS Sentiment Using TextBlob

```python

def find_Sentiment(val):                                                                                                               
    if val<=0.1 and val>-0.1:
        return 'Neutral'
    elif val>0.1:
        return 'Positive'
    else:
        return 'Negative'

def newsAnalysis(articles): 
    sent = []
    sentVal = []
    subjectivity = []
    for article in articles.iteritems():
        analysis = TextBlob(article[1])
        sentiment = find_Sentiment(analysis.sentiment.polarity)
        polarity = analysis.sentiment.polarity
        subjective = analysis.sentiment.subjectivity
        sentVal.append(polarity)
        sent.append(sentiment)
        subjectivity.append(subjective)
    FOX_Sentiment['polarity']=sentVal
    FOX_Sentiment['sentiment']=sent
    FOX_Sentiment['subjectivity']=subjectivity
    
FOX_Sentiment = pd.DataFrame(columns =['description','article_text','polarity','subjectivity','sentiment'])
FOX_Sentiment['article_text'] = FOX_data['article_text']
FOX_Sentiment['description'] = FOX_data['description']
newsAnalysis(FOX_data['article_text'])
FOX_Sentiment
```

## FOX NEWS Sentiment Using vaderSentiment

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python

def newsAnalysis_vader(articles): 
    scores = []

    for article in articles.iteritems():
        analysis = SentimentIntensityAnalyzer()
        vs = analysis.polarity_scores(article[1])
        scores.append(vs)
        vader_sent = pd.DataFrame(scores)
    return vader_sent


def find_Sentiment(vader_sentiment):
    Sentiment = []
    for index,row in vader_sentiment.iterrows():
        if row['vader_compound'] < -0.6 and row['vader_neg']-row['vader_pos'] > 0 :
            Sentiment.append('Negative')
        elif row['vader_compound'] > 0.6 and row['vader_pos']-row['vader_neg'] > 0 :
            Sentiment.append('Positive')
        else:
            Sentiment.append('Neutral')
    return Sentiment
    
vader_sentiment = newsAnalysis_vader(FOX_data['article_text'])
vader_sentiment.columns=['vader_compound','vader_neg','vader_new','vader_pos']

Sentiment = find_Sentiment(vader_sentiment)
vader_sentiment['Vader_Sentiment'] = Sentiment
vader_sentiment
FOX_Sentiment_New = pd.concat([FOX_Sentiment,vader_sentiment],axis=1)
FOX_Sentiment_New
```

##  Most Common Postive Words  and Negative Words

```python
import nltk as nlp
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
nlp.download('stopwords')
```

##  FOX News Most common words

```python
token = RegexpTokenizer(r'[a-zA-Z]+')
stop_words = stopwords.words('english')
def text_processing(articletext):
    tokens = list()
    for text in articletext:
        toks = token.tokenize(text)
        toks = [t.lower() for t in toks if t.lower() not in stop_words]
        tokens.extend(toks)    
    return tokens
positive_texts = FOX_Sentiment_New[FOX_Sentiment_New.Vader_Sentiment == "Positive"].article_text
positive_tokens = text_processing(positive_texts)
positive_freq = nlp.FreqDist(positive_tokens)
positive_freqlist = positive_freq.most_common(20)
positive_words_df = pd.DataFrame(positive_freqlist,columns=['Positive_words','Pos_word_freq'])
negative_texts = FOX_Sentiment_New[FOX_Sentiment_New.Vader_Sentiment == "Negative"].article_text
negative_tokens = text_processing(negative_texts)
negative_freq = nlp.FreqDist(negative_tokens)
negative_freqlist = negative_freq.most_common(20)
negative_words_df = pd.DataFrame(negative_freqlist,columns=['Negative_words','Neg_word_freq'])
FOX_words_df = pd.concat([positive_words_df,negative_words_df],axis=1)
FOX_words_df
```

## CNN News Most common words

```python
def cnn_text_processing(articletext):
    tokens = list()
    for text in articletext:
        toks = token.tokenize(text)
        toks = [t.lower() for t in toks if t.lower() not in stop_words]
        tokens.extend(toks)    
    return tokens
cnn_positive_texts = CNN_Sentiment_New[CNN_Sentiment_New.Vader_Sentiment == "Positive"].article_text
cnn_positive_tokens = cnn_text_processing(cnn_positive_texts)
cnn_positive_freq = nlp.FreqDist(cnn_positive_tokens)
cnn_positive_freqlist = cnn_positive_freq.most_common(20)
cnn_positive_words_df = pd.DataFrame(cnn_positive_freqlist,columns=['Positive_words','Pos_word_freq'])
cnn_negative_texts = CNN_Sentiment_New[CNN_Sentiment_New.Vader_Sentiment == "Negative"].article_text
cnn_negative_tokens = text_processing(cnn_negative_texts)
cnn_negative_freq = nlp.FreqDist(cnn_negative_tokens)
cnn_negative_freqlist = cnn_negative_freq.most_common(20)
cnn_negative_words_df = pd.DataFrame(cnn_negative_freqlist,columns=['Negative_words','Neg_word_freq'])
CNN_words_df = pd.concat([cnn_positive_words_df,cnn_negative_words_df],axis=1)
CNN_words_df
```
